```python
from google import genai
from google.genai import types
import base64

def generate():
  client = genai.Client(
      vertexai=True,
      project="qwiklabs-gcp-00-2096e75d3690",
      location="us-east4",
  )

  msg1_image1 = types.Part.from_uri(
      file_uri="gs://qwiklabs-gcp-00-2096e75d3690-labconfig-bucket/cymbal-product-image.png",
      mime_type="image/png",
  )
  msg1_text1 = types.Part.from_text(text="""Analyze the attached Cymbal Direct product image and generate three distinct descriptions:

1. A short, factual product description (1–2 lines).
2. A catchy advertising phrase suitable for a product website or banner.
3. A poetic, nature-themed description that evokes emotion and connection to the outdoors.

Be vivid, expressive, and creative. Keep tone appropriate for a young, outdoorsy audience.""")
  msg2_text1 = types.Part.from_text(text="""Okay, here are three distinct descriptions for the image of the backpack in the wildflower meadow, tailored for Cymbal Direct's young, outdoorsy audience:

**1. Short, Factual Product Description:**

\"Hit the trails prepared. This durable backpack offers ample space and comfortable support for all your adventures, from day hikes to weekend expeditions.\"

**2. Catchy Advertising Phrase:**

\"Unleash Your Wanderlust: Pack up, step out, and explore. Adventure awaits, are you Ready?\"

**3. Poetic, Nature-Themed Description:**

\"Where the wildflower tapestry meets the winding path, your journey unfolds with every step. This trusty pack, your steadfast companion, bearing the weight of dreams against a backdrop of blooming freedom.\"""")
  msg3_text1 = types.Part.from_text(text="""Using the attached Cymbal Direct product image, generate three unique outputs:

1. A vivid product description in under 10 words.  
2. A bold, creative advertising phrase (max 10 words).  
3. A poetic, unexpected nature-themed line (under 10 words).  

Make each line wildly creative, expressive, and emotionally engaging. Use unusual, imaginative phrasing where possible. Surprise the reader."""
)
  msg4_text1 = types.Part.from_text(text="""Okay, I'm ready! Please provide the image. I will then generate the three descriptions as requested: a short factual description, a catchy advertising phrase, and a poetic, nature-themed description.""")
  msg5_image1 = types.Part.from_uri(
      file_uri="gs://qwiklabs-gcp-00-2096e75d3690-labconfig-bucket/cymbal-product-image.png",
      mime_type="image/png",
  )
  msg5_text1 = types.Part.from_text(text="""Analyze the attached Cymbal Direct product image and generate three distinct descriptions:

1. A short, factual product description (1–2 lines).
2. A catchy advertising phrase suitable for a product website or banner.
3. A poetic, nature-themed description that evokes emotion and connection to the outdoors.

Be vivid, expressive, and creative. Keep tone appropriate for a young, outdoorsy audience.""")
  si_text1 = """You are helping Cymbal Direct analyze their product images to generate high-quality marketing content. Focus on capturing the visual, emotional, and thematic essence of the product in various creative styles, including short descriptions, ad-friendly phrases, and poetic language."""

  model = "gemini-2.0-flash-001"
  contents = [
    types.Content(
      role="user",
      parts=[
        msg1_image1,
        msg1_text1
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg2_text1
      ]
    ),
    types.Content(
      role="user",
      parts=[
        msg3_text1
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg4_text1
      ]
    ),
    types.Content(
      role="user",
      parts=[
        msg5_image1,
        msg5_text1
      ]
    ),
  ]

  generate_content_config = types.GenerateContentConfig(
    temperature = 2,
    top_p = 1,
    max_output_tokens = 3627,
    safety_settings = [types.SafetySetting(
      category="HARM_CATEGORY_HATE_SPEECH",
      threshold="OFF"
    ),types.SafetySetting(
      category="HARM_CATEGORY_DANGEROUS_CONTENT",
      threshold="OFF"
    ),types.SafetySetting(
      category="HARM_CATEGORY_SEXUALLY_EXPLICIT",
      threshold="OFF"
    ),types.SafetySetting(
      category="HARM_CATEGORY_HARASSMENT",
      threshold="OFF"
    )],
    system_instruction=[types.Part.from_text(text=si_text1)],
  )

  for chunk in client.models.generate_content_stream(
    model = model,
    contents = contents,
    config = generate_content_config,
    ):
    print(chunk.text, end="")

generate()
```
